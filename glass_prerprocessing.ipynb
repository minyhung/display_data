```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# 원하는 피처 선택
selected_features = ['R_RD', 'L_RD', 'L_Vac', 'R_Vac']

# 선택한 피처들을 포함한 데이터 추출
selected_data = result_data[selected_features]

# 연관규칙 분석을 위한 더미 변수 생성
dummy_data = pd.get_dummies(selected_data, dummy_na=False)  # '0' 값을 'False'로 변환

# 지지도 계산
frequent_itemsets = apriori(dummy_data, min_support=0.05, use_colnames=True)

# 연관규칙 생성
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.5) 


# 추가 분석을 위한 지지도, 신뢰도, lift, position, scale_lift 계산
rules['position'] = rules['antecedents'].apply(lambda x: len(x))
rules['scale_lift'] = rules['lift'] / rules['position']

# 결과를 데이터프레임 형태로 출력
print("Association Rules:")
print(rules)

print("\nAdditional Analysis:")
additional_analysis = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift', 'position', 'scale_lift']]
print(additional_analysis)

```

```python
import pandas as pd
from sklearn.feature_selection import RFE
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# 데이터 불러오기
result_data = pd.read_csv("C:/Users/mhlee/OneDrive/바탕 화면/엘지 헬로비전 DX 데이터스쿨/LG 디스플레이 공정데이터/after_ROLL_VAC_data.csv", encoding='cp949', index_col=0)

# 독립변수와 종속변수 분리
X = result_data.dropna()  # 결측치가 있는 행 삭제
y = X["L_RD"]  # 종속변수 설정
X = X.drop(columns=["L_RD"])  # 종속변수 제거

# RFE를 사용하여 50개의 독립변수 중에서 5개를 선택
xgb = XGBRegressor()  # XGBoost 모델 생성
rfe = RFE(estimator=xgb, n_features_to_select=5)  # RFE 모델 생성
rfe.fit(X, y)  # RFE 모델 학습

# 선택된 5개의 변수 확인
selected_features = X.columns[rfe.support_]
print("Selected Features:", selected_features)

# 선택된 5개의 변수로만 이루어진 데이터프레임 생성
X_selected = X[selected_features]

# Train/Test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# XGBoost 모델 훈련
model = XGBRegressor()
model.fit(X_train, y_train)

# 테스트 데이터에 대한 예측
y_pred = model.predict(X_test)

# 모델 평가 (예시로 RMSE 사용)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("RMSE:", rmse)
```
