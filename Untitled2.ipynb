```python
# 패키지 임포트
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
```

```python

display_data = pd.read_csv('factory_glass_2016.csv', encoding='cp949',index_col=0)
```

```python
display_data.head()
```

```python
# null 값 확인하기

#결측치가 많은 값을 먼저 확인하고 싶어서, 오름차순으로 진행

null_value = display_data.isnull().sum().sort_values(ascending=False)
print(null_value)

```

```python
# 조회 결과, 7개 컬럼은 전체 8145개의 행이 모두 결측치. 행 전체 삭제

print(null_value[:20])

     
```

```python
# 전체 행이 결측치인 컬럼을 찾고 삭제하기 #7개 삭제

all_null_columns = null_value[null_value == len(display_data)].index
#print(all_null_columns)

# 해당 열들을 데이터프레임에서 삭제
df_cleaned = display_data.drop(columns=all_null_columns)

print(df_cleaned)     
```

```python
#선형 보간법 사용해서 데이터 결측치 채우기 
df_interpolated = df_cleaned.interpolate(method='linear', axis=0)

has_null_values = df_interpolated.isnull().any()

# 결과 출력
print("결측치 여부 확인:", has_null_values)
```

```python
#분산이 0인 피처 제거

#분산이 0인 것은 해당 피처가 모든 값이 동일한 값을 가지고 있음을 의미
# 즉, 모든 값이 동일하기 때문에 피처의 변동이 없음

# 모델이 해당 피처를 이용하여 예측을 수행하는 것은 의미가 없습니다. 
#왜냐하면 모든 값이 같으므로 피처가 예측에 어떠한 영향을 미치지 않기 때문입니다

#모델 성능 높이는데 도움
```

```python
variances = df_interpolated.var()

# 분산이 0에 가까운 피처의 인덱스를 찾기
zero_variance_features = variances[variances <= 10e-10].index

# 분산이 0인 피처를 데이터프레임에서 제거
df_filtered = df_interpolated.drop(columns=zero_variance_features)

# 결과 출력
print("분산이 0인 피처를 제거한 후의 데이터프레임 크기:", df_filtered.shape)
```

```python
# 모든 컬럼에 대해 상관계수의 절대값 구하기
corr_arr = abs(df_filtered.corr(method='pearson'))
```

```python
# 상관계수의 절대값 히트맵 확인
plt.figure(figsize=(20,20))
sns.heatmap(corr_arr)

# 히트맵을 확인해보면, 상관계수가 높은 피처를 쉽게 확인할 수 있음
```

```python
#종속성 높은 피처 값 구하고 제거
# 상관 행렬 계산
correlation_matrix = df_interpolated.corr().abs()

# 상관 계수가 높은 피처 쌍을 찾기
high_correlation_features = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if correlation_matrix.iloc[i, j] >= 0.9:
            colname = correlation_matrix.columns[i]
            high_correlation_features.add(colname)

# 상관 계수가 높은 피처 제거
df_filtered = df_interpolated.drop(columns=high_correlation_features)

# 결과 출력
print("상관 관계가 높은 피처를 제거한 후의 데이터프레임 크기:", df_filtered.shape)  #352개
```

```python
#VIF 활용해서 다중공선성 처리

from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(df):
    # VIF 계산
    vif_data = pd.DataFrame()
    vif_data["feature"] = df.columns
    vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(len(df.columns))]
    return vif_data

# VIF 계산
vif_result = calculate_vif(df_interpolated)

# VIF가 30 이상인 피처를 제거
high_vif_features = vif_result[vif_result["VIF"] >= 30]["feature"].tolist()
df_filtered = df_interpolated.drop(columns=high_vif_features)

# 결과 출력
print("VIF가 30 이상인 피처를 제거한 후의 데이터프레임 크기:", df_filtered.shape)  #53개
```
